```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]

def forward(x):
    return x*w+b

def loss(x,y):
    y_pred = forward(x)
    return (y_pred - y) * (y_pred - y)

w_list = []
b_list = []
mse_list = []

for w in np.arange(0.0,4.1,0.1):
    for b in np.arange(-4.0,4.1,0.1):
        print('w=',w,'b=',b)
        l_sum = 0
        for x_val,y_val in zip(x_data,y_data):
            y_prel_val = forward(x_val)
            loss_val = loss(x_val,y_val)
            l_sum += loss_val
            print('\t',x_val,y_val,y_prel_val,loss_val)
        print('MSE=',l_sum / 3)
        w_list.append(w)
        b_list.append(b)
        mse_list.append(l_sum / 3)
        
```

```python
fig = plt.figure() 
ax3 = plt.axes(projection='3d')
ax3.set_xlim(-4.0,4.0)
ax3.set_ylim(-4.0,4.0)

xx = w_list
yy = b_list
X, Y = np.meshgrid(xx, yy)
Z = np.sin(X)+np.cos(Y)
ax3.plot_surface(X,Y,Z,cmap='rainbow')
#ax3.contour(X,Y,Z, zdim='z',offset=0，cmap='rainbow')   #等高线图，要设置offset，为Z的最小值
plt.show()
```
